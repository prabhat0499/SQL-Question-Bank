* Find the customers with the highest daily total order cost between 2019-02-01 and 2019-05-01.

## Solution

```python
%%sql
WITH daily_costs as (
    select
        c.first_name,
        o.order_date,
        sum(o.total_order_cost) as daily_total
    from customers c
    join orders o
    on c.id=o.cust_id
    where order_date between '2019-02-01' and '2019-05-01'
    group by c.first_name,o.order_date
),
ranked as (
    select *,
        row_number() over (partition by order_date order by daily_total desc) as rk
    from daily_costs
)
select * from ranked
where rk=1
```

* ROW_NUMBER is dropping rows when there are ties (i.e., multiple customers with the same maximum daily spend).
* For example:
    * On 2019-03-01, both Mia (80) and Farida (80) should appear.
    * On 2019-03-11, both Mia (20) and Eva (20) should appear.
    * On 2019-04-19, both Jill (275) and Mark (275) should appear.
------------
* But your query using ROW_NUMBER() only keeps one row per day.

* To fix this, you should use RANK() (or DENSE_RANK()) instead of ROW_NUMBER() so ties are preserved.

```python
%%sql
WITH daily_costs as (
    select
        c.first_name,
        o.order_date,
        sum(o.total_order_cost) as daily_total
    from customers c
    join orders o
    on c.id=o.cust_id
    where order_date between '2019-02-01' and '2019-05-01'
    group by c.first_name,o.order_date
),
ranked as (
    select *,
        rank() over (partition by order_date order by daily_total desc) as rk
    from daily_costs
)
select * from ranked
where rk=1
```

* See, here - On 2019-03-01, both Mia (80) and Farida (80) appears
