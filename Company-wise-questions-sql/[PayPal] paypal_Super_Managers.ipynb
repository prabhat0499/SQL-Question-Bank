* Find managers with at least 7 direct reporting employees. In situations where user is reporting to himself/herself, count that also.
Output first names of managers.

### Method 1 - Longer method on the first attempt, this is a HARD problem

```python
%%sql
select 
    e.id as emp_id,
    e.first_name as employee_name,
    m.first_name as manager_name
from employees e
join employees m
on e.manager_id=m.id
order by emp_id
```

```python
%%sql
with managers as (
    select 
        e.id as emp_id,
        e.first_name as employee_name,
        m.first_name as manager_name
    from employees e
    join employees m
    on e.manager_id=m.id
    order by emp_id
),
temp as (
    select 
        manager_name,
        employee_name,
        emp_id,
        row_number() over (partition by manager_name order by emp_id) as rk
    from managers   
) 
select manager_name, count(rk) as count_of_reportees
from temp
group by manager_name
having count_of_reportees>=7
```

```python
%%sql
with managers as(
    select
        m.first_name as manager_name,
        count(*) as count_of_employees
    from employees e
    join employees m
    on e.manager_id=m.id
    group by manager_name
    having count_of_employees>=7
    
)
select manager_name
from managers
```
